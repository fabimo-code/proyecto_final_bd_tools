## 2. Cargar módulos del proyecto

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))

from bdtools.io import read_raw_csv, run_etl
from bdtools.eda import run_eda
from bdtools.modeling import train_high_impact_classifier, run_spatial_clustering
from bdtools.config import PATHS

PATHS
```

## 4. Limpieza y exportación

Genera `data/processed/sievcac_clean.parquet` y tablas de calidad.

```python
df, quality = run_etl(PATHS.raw_csv)
print(df.shape)
quality.head(10)
```

## 5. Validaciones rápidas

```python
display(df[['id_caso','anio','mes_original','dia_original','precision_fecha','fecha_aproximada','departamento','municipio','latitud','longitud','total_de_victimas_del_caso','alto_impacto']].head())
print(df['anio'].min(), df['anio'].max())
print(df['alto_impacto'].value_counts(normalize=True))
```

## 6. EDA inicial

```python
outputs = run_eda(df)
outputs
```

## 7. Modelo inicial: alto impacto
